# Q31: Visualizations for Correlations and Feature Importance

- Combine Pandas with Matplotlib/Seaborn to create insightful visualizations (heatmaps, scatter plots, pair plots).
- Use them to explain correlations and feature importance in a dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
```

```python
# Load dataset
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['MedHouseVal'] = housing.target
```

```python
# --- Correlation Heatmap ---
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# --- Scatter Plot: MedInc vs MedHouseVal ---
plt.figure(figsize=(7,5))
sns.scatterplot(x='MedInc', y='MedHouseVal', data=df)
plt.title('Median Income vs Median House Value')
plt.show()
```

```python
# --- Pair Plot ---
sns.pairplot(df[['MedInc', 'AveRooms', 'AveOccup', 'MedHouseVal']])
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()
```

```python
# --- Explanation ---
print('The heatmap shows strong positive correlation between MedInc and MedHouseVal. Scatter and pair plots reveal feature relationships and help identify important predictors.')
```
